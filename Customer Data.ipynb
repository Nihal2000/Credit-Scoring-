```python
#load raw data
raw_data= pd.read_csv("data.csv", encoding="ISO-8859-1")
```

```python
#make a dataframe for date time
date_time= raw_data['InvoiceDate'].str.split(" ", expand= True)
date_time.head()
```

```python
data= raw_data
data['Time']= date_time[1]
```

```python
data.head()
```

```python
date= date_time[0].str.split("/", expand= True)
data['Month'], data['Date'], data['Year']= date[0], date[1], date[2]
```

```python
#delete null values
data= data[~data.CustomerID.isna()]
data.head()
```

```python
#data.head()
data.shape
```

```python
#delete last month of 2011
data= data[((data['Year']!='2011') | (data['Month']!='12'))]
```

```python
#make Invoice Timestamp
data['InvoiceDate']= pd.to_datetime(data['InvoiceDate'])
data.tail()
```

```python
#data.head()
data.shape
```

```python
#unique data Customer id
Customer_ID= data['CustomerID'].unique()
Customer_ID.shape
```

```python
#Price
data['Price']= data['Quantity']* data['UnitPrice']
data.head()
```

```python
#for all customer

Customer_Score= pd.DataFrame(columns= ['CustomerID', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
point=1

for customer in Customer_ID:
    data_customer= data[data.CustomerID==customer]
    customer_month_group= data_customer.groupby(data_customer['InvoiceDate'].dt.strftime('%b'))['Price', 'Quantity'].sum()
    customer_month_group= customer_month_group.reset_index()
    customer_month_group['InvoiceDate'] = pd.to_datetime(customer_month_group.InvoiceDate, format='%b', errors='coerce').dt.month
    customer_month_group = customer_month_group.sort_values(by='InvoiceDate')
    customer_month_group= customer_month_group.reset_index()
    customer_month_group= customer_month_group.drop(['index'], axis=1)
    
    unique_date= data_customer.InvoiceDate.dt.date.unique()
    
    date1= pd.to_datetime(unique_date)
    date1= date1.groupby(date1.year)
    
    count = {}

    for key, values in date1.items():
        values= pd.to_datetime(values)
        values= values.groupby(values.month)
        for key, values in values.items():
            count[key]= len(values)

    add= []
    add.insert(0, customer)
    month=1
    for index, row in customer_month_group.iterrows(): 
        #print (row['InvoiceDate'], row['Price'], row['Quantity'])
        i, price, quantity = row['InvoiceDate'], row['Price'], row['Quantity']
        #print(i, price, quantity, count[i])
        while(month<i and month < 13):
            add.insert(int(month), [0, 0, 0])
            month= month+1
        add.insert(int(i), [price, quantity, count[i]])
        month= month+1
    
    while(month<13):
        add.insert(int(month), [0, 0, 0])
        month= month+1

    Customer_Score.loc[point]= add
    point= point+1
```
