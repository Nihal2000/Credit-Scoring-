```python
import numpy as np
import pandas as pd
```

```python
Customer_Data= pd.read_csv('Customer_Data1.csv')
Customer_Median_Data= pd.read_csv('Customer_Median_Data1.csv')
```

```python
#set a customer
customer= 17841
```

```python
#set current month
current_month= 11
```

```python
#create a new dataframe for saving Score for one customer
Customer_Score_Data= pd.DataFrame(columns= ['CustomerID', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
```

```python
#create a new dataframe for saving lastest Score and median
Customer_Score_Median= pd.DataFrame(columns= ['CustomerID', 'Price', 'Quantity', 'Freq', 'Score'])
```

```python
Customer_Score_Median
```

```python
customer_data= Customer_Data[Customer_Data.CustomerID==customer]
customer_median_data= Customer_Median_Data[Customer_Median_Data.CustomerID==customer]
```

```python
def ind_Score(val, med):
    if(val==med):
        return 0.0;
    if(val>med):
        if(val>(med*1.1)):
            return 0.75
        if(val>(med*1.15)):
            return 0.1
        return 2.0
    if(val<med):
        if(val>(med*0.9)):
            return -0.15
        if(val>(med*0.85)):
            return -0.5
        if(val>(med*0.75)):
            return -1.0
        return -1.5
```

```python
month= (current_month+1)%13
if(month==0):
    month+=1
add= []
add.insert(0, customer)
Customer_Score_Median.loc[69]= [17841, 0.0, 0.0, 0.0, 0.0]

for i in range(1, 13):
    customer_price= float(customer_data.loc[69][month].strip('][').split(', ')[0])
    customer_quan= float(customer_data.loc[69][month].strip('][').split(', ')[1])
    customer_frq= float(customer_data.loc[69][month].strip('][').split(', ')[2])
    
    price_sc= ind_Score(customer_price, Customer_Score_Median.loc[69][1])
    quan_sc= ind_Score(customer_quan, Customer_Score_Median.loc[69][2])
    frq_sc= ind_Score(customer_frq, Customer_Score_Median.loc[69][3])
    
    s_Score= (0.4*price_sc) + (0.35*frq_sc) + (0.25*quan_sc)
    Score= Customer_Score_Median.loc[69][4]+s_Score
    
    print(month, s_Score, Score, Customer_Score_Median.loc[69][4])
    
    Customer_Score_Median.loc[69][4]= Score
    
    Customer_Score_Median.loc[69][1]= float(customer_median_data.loc[69][month].strip('][').split(', ')[0])
    Customer_Score_Median.loc[69][2]= float(customer_median_data.loc[69][month].strip('][').split(', ')[1])
    Customer_Score_Median.loc[69][3]= float(customer_median_data.loc[69][month].strip('][').split(', ')[2])
    
    add.insert(int(month), Score)
    month= (month+1)%13
    if(month==0):
        month+= 1
Customer_Score_Data.loc[69]= add 
```

```python
#for all customer
current_month= 11
Customer_Score_Data= pd.DataFrame(columns= ['CustomerID', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
Customer_Score_Median= pd.DataFrame(columns= ['CustomerID', 'Price', 'Quantity', 'Freq', 'Score'])
CustomerID= Customer_Data['CustomerID'].unique()
point= 0

for customer in CustomerID:
    customer_data= Customer_Data[Customer_Data.CustomerID==customer]
    customer_median_data= Customer_Median_Data[Customer_Median_Data.CustomerID==customer]
    
    month= (current_month+1)%13
    if(month==0):
        month+=1
    add= []
    add.insert(0, customer)
    Customer_Score_Median.loc[point]= [customer, 0.0, 0.0, 0.0, 0.0]

    for i in range(1, 13):
        customer_price= float(customer_data.loc[point][month].strip('][').split(', ')[0])
        customer_quan= float(customer_data.loc[point][month].strip('][').split(', ')[1])
        customer_frq= float(customer_data.loc[point][month].strip('][').split(', ')[2])

        price_sc= ind_Score(customer_price, Customer_Score_Median.loc[point][1])
        quan_sc= ind_Score(customer_quan, Customer_Score_Median.loc[point][2])
        frq_sc= ind_Score(customer_frq, Customer_Score_Median.loc[point][3])

        s_Score= (0.4*price_sc) + (0.35*frq_sc) + (0.25*quan_sc)
        Score= Customer_Score_Median.loc[point][4]+s_Score

        #print(month, s_Score, Score, Customer_Score_Median.loc[point][4])

        Customer_Score_Median.loc[point][4]= Score

        Customer_Score_Median.loc[point][1]= float(customer_median_data.loc[point][month].strip('][').split(', ')[0])
        Customer_Score_Median.loc[point][2]= float(customer_median_data.loc[point][month].strip('][').split(', ')[1])
        Customer_Score_Median.loc[point][3]= float(customer_median_data.loc[point][month].strip('][').split(', ')[2])

        add.insert(int(month), Score)
        month= (month+1)%13
        if(month==0):
            month+= 1
    Customer_Score_Data.loc[point]= add 
    point += 1
```
