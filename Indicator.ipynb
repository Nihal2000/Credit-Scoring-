```python
import numpy as np
import pandas as pd
```

```python
Customer_Data= pd.read_csv("Customer_Data1.csv")
Customer_Median_Data= pd.read_csv("Customer_Median_Data1.csv")
```

```python
#set a customer
customer= 17841
```

```python
#set current month
current_month= 11
```

```python
#create a new dataframe for setting indicator for one customer
Customer_Indicator= pd.DataFrame(columns= ['CustomerID', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
```

```python
#retrive data from Customer_Data and Customer_Median_Data
customer_data= Customer_Data[Customer_Data.CustomerID==customer]
customer_median_data= Customer_Median_Data[Customer_Median_Data.CustomerID==customer]
```

```python
#create indication for one customer
month= (current_month+1)%13
add_data= []
add_data.insert(0, customer)

def indicator(data, median):
    data= round(data, 2)
    median= round(median, 2)
    if(data>median):
        return 1
    if(data==median):
        return 0
    if(data<median):
        return -1

for i in range(1,13):
    customer_price= float(customer_data.loc[69][month].strip('][').split(', ')[0])
    customer_quan= float(customer_data.loc[69][month].strip('][').split(', ')[1])
    customer_frq= float(customer_data.loc[69][month].strip('][').split(', ')[2])
    customer_median_price= float(customer_median_data.loc[69][month].strip('][').split(', ')[0])
    customer_median_quan= float(customer_median_data.loc[69][month].strip('][').split(', ')[1])
    customer_median_frq= float(customer_median_data.loc[69][month].strip('][').split(', ')[2])
    add_data.insert(int(month), [indicator(customer_price, customer_median_price), indicator(customer_quan, customer_median_quan),indicator(customer_frq, customer_median_frq)])
    month= (month+1)%13
Customer_Indicator.loc[0]= add_data
```

```python
#create indication for all customer
Customer_Indicator= pd.DataFrame(columns= ['CustomerID', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
current_month= 11
point= 0
customerID= Customer_Data['CustomerID'].unique()

def indicator(data, median):
        data= round(data, 2)
        median= round(median, 2)
        if(data>median):
            return 1
        if(data==median):
            return 0
        if(data<median):
            return -1

for customer in customerID:

    month= (current_month+1)%13
    add_data= []
    add_data.insert(0, customer)
    customer_data= Customer_Data[Customer_Data.CustomerID==customer]
    customer_median_data= Customer_Median_Data[Customer_Median_Data.CustomerID==customer]
    
    for i in range(1,13):
        customer_price= float(customer_data.loc[point][month].strip('][').split(', ')[0])
        customer_quan= float(customer_data.loc[point][month].strip('][').split(', ')[1])
        customer_frq= float(customer_data.loc[point][month].strip('][').split(', ')[2])
        customer_median_price= float(customer_median_data.loc[point][month].strip('][').split(', ')[0])
        customer_median_quan= float(customer_median_data.loc[point][month].strip('][').split(', ')[1])
        customer_median_frq= float(customer_median_data.loc[point][month].strip('][').split(', ')[2])
        add_data.insert(int(month), [indicator(customer_price, customer_median_price), indicator(customer_quan, customer_median_quan),indicator(customer_frq, customer_median_frq)])
        month= (month+1)%13
        if(month==0):
            month+= 1
    Customer_Indicator.loc[point]= add_data
    point= point+1
```
